```python
#Part 2

dic={'one':'1','two':'2','three':'3','four':'4','five':'5','six':'6','seven':'7','eight':'8','nine':'9'}

def find_first_digit(line):
    
    """Finds the first digit, or by recognizing the first number, or recognizing the first string"""
    
    #Finding the first index of number i_num
    i_num=float("inf")
    for i in range(len(line)):
        if line[i].isnumeric():
            i_num=i
            break
    
    k=i_num
    
    #Now finding the first index of string
    good_key=''
    for i in range(len(line)):
        for key in dic.keys():
            if key in line[:i]:
                m=len(key)
                if i-m<k:
                    k=i-m
                    good_key=key
                    break
    
    #If first is number, return number, else return number value of the string
    if k==i_num:
        return(line[k])
    else:
        return(dic[good_key])


def find_last_digit(line):
    """Same but for finding last digit"""
    i_num=0
    for i in range(len(line)):
        if line[i].isnumeric():
            i_num=i
                
                
    k=i_num   
    good_key=''
    for key in dic.keys():
        r=find_last_index(key,line)
        if r>k:
            k=r
            good_key=key
                
    if k==i_num:
            
        digit2=line[i_num]
        
    else:
        digit2=dic[good_key]
        
    return(digit2)


def find_last_index(string,txt):
    """Find the index where the last number in string format begins"""
    k=0
    for i in range(len(txt)):
        #print(txt[i:])
        #print(i)
        if string in txt[i:]:
            k=i
    return(k)


def find_number(line):

    
    """Assemblate first and second digit"""

    digit1=find_first_digit(line)
    digit2=find_last_digit(line)        
    digit=digit1+digit2
        
    return(digit)




#Pb resolution:
k=0

#Reading file
with open('input.txt','r') as f:
    lines=f.readlines()
    data=[line.strip() for line in lines]
    
    #Parcours des lignes
    for line in data:

        number=find_number(line)
        k+=int(number)
    
print(k)  
```
