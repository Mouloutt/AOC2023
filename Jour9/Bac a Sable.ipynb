```python
def diff_seq(L):
    return([L[i+1]-L[i] for i in range(len(L)-1)])
    
def get_seq_list(L):
    A=[L]
    K=L
    while not set(K).issubset(set([0])):
        K=diff_seq(K)
        A.append(K)
    return(A)

def find_last_elt(L):
    A=get_seq_list(L)[::-1]
    #print(A)
    for i in range(1,len(A)):
        A[i].append(A[i][-1]+A[i-1][-1])
        #print(A[i])
    return(A[-1][-1])
```

```python
def find_first_elt(L):
    A=get_seq_list(L)[::-1]
    A=[L[::-1] for L in A]
    #print(A)
    for i in range(1,len(A)):
        A[i].append(A[i][-1]-A[i-1][-1])
        #print(A[i])
    return(A[-1][-1])
```

```python
f= open('input.txt','r')
lines=[line.strip() for line in f.readlines()]
lines=[[int(i) for i in line.split()] for line in lines]
k=0
for line in lines:
    k+=find_first_elt(line)
print(k)
```
