```python
dic={v:str(v) for v in range(2,15)}
dic[11]='J'
dic[10]='T'
dic[12]="Q"
dic[13]="K"
dic[14]='A'
dic={v:k for k,v in dic.items()}

def get_power(h):
    max_occ=max(h,key=lambda x:h.count(x))
    return(h.count(max_occ),max_occ)
def get_rank(h):
    power,max_occ=get_power(h)
    if power==1:
        return(1)
    elif power in [4,5]:
        return(power+2)
    elif power==3:
        h2=h.replace(max_occ,'')
        power2,max_occ2=get_power(h2)
        if power2==2:
            return(5)
        else:
            return(4)
    elif power==2:
        h2=h.replace(max_occ,'')
        power2,max_occ2=get_power(h2)
        if power2==2:
            return(3)
        else:
            return(2)
            
def get_points(h):
    return([get_rank(h)]+[dic[i] for i in h])

def compare_tie(h1,h2):
    i = next((i for i, (char1, char2) in enumerate(zip(h1, h2)) if char1 != char2), None)
    return([h1,h2][dic[h2[i]]>dic[h1[i]]])
```
