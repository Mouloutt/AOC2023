```python
f= open('input.txt','r')


maps=[line.strip() for line in f.readlines()]
```

```python
def full_number(v,i,k):
    global maps
    x,y=v
    if i==0:
        
    
        return(int(maps[x][y:y+k]))
    elif i==1:
        return(int(maps[x][y-1:y+k-1]))
        
    elif i==2:
        return(int(maps[x][y-2:y+k-2]))

def get_number_length(v):
    x,y=v
    global maps
    L=maps[x][y:].split('.')
    c=0
    for car in L[0]:
        if car.isdigit():
            c+=1
        else:
            break
    return(c)
    
def find_neighbours(v):
    x,y=v

    return((x,y-1),(x,y+1),(x+1,y-1),(x+1,y+1),(x-1,y+1),(x-1,y-1),(x+1,y),(x-1,y))
    
    
```

```python
def find_special_car(v):
    global maps
    answer=False
    for s in find_neighbours(v):
        x,y=s
        try:
            if maps[x][y] not in '0123456789.':
                answer=True
        except:
            pass
    return(answer)
        
```

```python
find_special_car((0,12))
```

```python
k=0
for line in maps:
    i=maps.index(line)
    iterator=iter(range(len(line)))
    for j in iterator:
        #print('scanning caract {} in line {}'.format(j,i))
        if line[j].isdigit():
            print('digit found at position ({},{})'.format(i,j))
            l=get_number_length((i,j))
            print('it is part of a {}-digits number'.format(l))

            for m in range(l):
                if find_special_car((i,j+m)):
                    print("special caract found for digit at position: {} for {}th caracter".format(m,j))
                    print("position of the digit where we look for number: ({},{})".format(i,j+m))
                    num=full_number((i,j+m),m,l)
                    print('number found: {}'.format(num))
                    k+=num
                    print('skipping {} iteration'.format(l))
                    for a in range(l):
                        try:
                            next(iterator)
                        
                        except:
                            pass
                    break
            else:
                print('No special caracter found around')


                        
                    

print(k)
                
                
```

```python
import re
def get_numbers(line):
    p=re.compile('[0-9]+')
    
    for number in p.finditer(line):
        yield number
        
def find_gear_ratio(i,j):
    L=[]
    global maps
    for k in range(max(0,i-1),min(len(maps),i+2)):
        for number in get_numbers(maps[k]):
            if (number.end()>=j) and (number.start()<=j+1):
                L.append(int(number.group()))
    print(len(L))
    if len(L)==2:
        return(L[0]*L[1])
    else:
        return(0)
        
        
                
```

```python
find_gear_ratio(0,50)
```

```python
with open('input.txt','r') as f:
    maps=f.read().splitlines()
    k=0
    for i in range(len(maps)):
        for j in range(len(maps[0])):
            if maps[i][j]=='*':
                k+=find_gear_ratio(i,j)
print(k)
```
